```python
import random
import time

class Fish:
    def __init__(self, grid, number_of_moves = 0):
        self.grid = grid
        self.number_of_moves = number_of_moves
        
    def movement(self):

        fish_position = []

        for row in range(len(self.grid)):
            fish_position.extend(
                (row, column)
                for column in range(len(self.grid[row]))
                if self.grid[row][column] == 't'
            )

        print("Initial Fish Positions:", fish_position)

        for element in fish_position:
            fish_row, fish_column = element

            if grid[fish_row][fish_column] == 't':
                
                direction = random.choice([(1, 0), (0, 1), (-1, 0), (0, -1)])  # right, down, left, up
                new_x, new_y = (fish_row + direction[0]) % len(grid), (fish_column + direction[1]) % len(grid[0])
                print("New position:", (new_x, new_y))
                
                if (0 <= new_x < len(grid)
                    and 0 <= new_y < len(grid[0])
                    and grid[new_x][new_y] != 't'
                    and grid[new_x][new_y] != 's'
                    and grid[new_x][new_y] == 0
                ):

                    grid[new_x][new_y] = 't'
                    event_1 = grid[new_x][new_y]
                    grid[fish_row][fish_column] = 0
                    self.number_of_moves += 1
                   
                    if event_1 and self.number_of_moves == 3:
                        grid[fish_row][fish_column] = 't'          
                        grid[new_x][new_y] = 't'
                        self.number_of_moves = 0       

        # Output the updated grid for debugging
        print("Updated Grid:")
        for line in grid:
            print(line)
        print("\n")

grid = [
    [0, 0, 0],
    [0, 't', 0],
    [0, 0, 0]
]

fish = Fish(grid)

for number_of_moves, _ in enumerate(range(6, 0, -1), start=1):
    print("Number of moves:", number_of_moves)
    fish.movement()

    time.sleep(1)
```
